```python
import solvers
import numpy as np
import matplotlib.pyplot as plt
```

```python
fun = lambda p,t,x: p*x + np.cos(t) - p*np.sin(t)

exact = lambda t, p: np.sin(t)

lambdas = [-10, -5]
fig, ax = plt.subplots(1,1)
x = np.linspace(0,5,500)
for i in lambdas:
    RK = solvers.Runge_Kutta(fun, [0,5], 0, 0.01, i)
    ax.plot(x, RK, label=f'R-K lambda {i}')
ax.plot(x, exact(x,i), label='Solucion exacata')
ax.grid()
ax.legend()
plt.show()
```

```python
p=1

fun = lambda p,t,x: p*np.exp(t)

exact = lambda t, p: p*np.exp(t)

fig, ax = plt.subplots(1,1)
x = np.linspace(0,1,100)
RK = solvers.Runge_Kutta(fun, [0,1], 1, 0.01, p)
ax.plot(x, RK, label=f'R-K')
ax.plot(x, exact(x,p), label='Solucion exacata')
ax.grid()
ax.legend()
plt.show()
```
